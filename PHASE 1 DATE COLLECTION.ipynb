# 1.The population of each state.

```python
#importing libraries
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
```

```python
#Connecting to the driver
driver=webdriver.Chrome(r'C:\Users\Admin\Downloads\chromedriver_win32\chromedriver.exe')

#Maximizing the automated
driver.maximize_window()

#opening the wikipedia website on automated chrome window
url=' https://en.wikipedia.org/wiki/List_of_states_in_India_by_past_population'
driver.get(url)
```

```python
#Extracting the names of states/UT
States_UT=[]
try:
    States_UT_tag=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr/td[2]/a')
    for i in States_UT_tag:
        States_UT.append(i.text)
except NoSuchElementException : #handling no such element exception
    States_UT.append('-')  
```

```python
#printing len
len(States_UT)
```

```python
#displaying states/UT
States_UT
```

```python
#Extracting the names of population 2001
population_2001=[]
try:
    population_2001_tag=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr/td[8]')
    for i in population_2001_tag:
        population_2001.append(i.text)
except NoSuchElementException : #handling no such element exception
    population_2001.append('-')  
```

```python
#printing len
len(population_2001)
```

```python
#displaying population_2001
population_2001[:36]
```

```python
#Extracting the names of population 2011
population_2011=[]
try:
    population_2011_tag=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr/td[9]')
    for i in population_2011_tag:
        population_2011.append(i.text)
except NoSuchElementException : #handling no such element exception
    population_2011.append('-')  
```

```python
#printing len
len(population_2011)
```

```python
#displaying population_2011
population_2011[:36]
```

```python
#Making DataFrame
Population_of_each_state=pd.DataFrame()
Population_of_each_state['State/UT']=States_UT
Population_of_each_state['Population(2001)']=population_2001
Population_of_each_state['Population(2011)']=population_2011
Population_of_each_state
```

```python
#Saving file to csv
Population_of_each_state.to_excel('Population.xlsx')
```
